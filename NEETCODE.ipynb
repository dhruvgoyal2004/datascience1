```python
class Solution(object):
    def isAnagram(self, s, t):
        a = sorted(s)
        b = sorted(t)
        if a == b:
            return True
        else:
            return False
#takes more time to execute bad time complexity o(nlogn)
```

```python
from collections import Counter
class Solution(object):
    def isAnagram(self, s, t):
        a = Counter(s)
        b = Counter(t)
        if a == b:
            return True
        else:
            return False
#takes log(n)
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        for i in range(0,len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i]+nums[j] == target:
                    return i,j #brute force n2
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        i = 0
        for i in range(0,len(nums)):
            a = target - nums[i]
            if a in nums[i+1:]:
                return i, nums.index(a,i+1) #best n worst n2
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        num_map = {}  
        for i, num in enumerate(nums):  
            complement = target - num
            if complement in num_map:  
                return [num_map[complement], i]
            num_map[num] = i  #O(n)
```
