```python
Key Differences:
+: Combines lists into one flat list.
zip(): Creates a list of tuples by pairing elements from both lists.
```

```python
#merge 2 list and convert into dict
name = ["aman", "arav", "dhruv", "rahul"]
contact = [1234, 5678, 91011, 121314]
book = dict(zip(name,contact))
print(book)
```

```python
class Solution(object):
    def isAnagram(self, s, t):
        a = sorted(s)
        b = sorted(t)
        if a == b:
            return True
        else:
            return False
#takes more time to execute bad time complexity o(nlogn)
```

```python
from collections import Counter
class Solution(object):
    def isAnagram(self, s, t):
        a = Counter(s)
        b = Counter(t)
        if a == b:
            return True
        else:
            return False
#takes log(n)
```
